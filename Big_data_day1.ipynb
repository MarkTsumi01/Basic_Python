```python
#026
def latin(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if word[0] in vowels:
        pig_word = word + 'way'
    else:
        pig_word = word[1:] + word[0] + 'ay'
    return pig_word.lower()

user_input = input("Enter a word: ")
piglatin_word = latin(user_input)
print("Pig Latin conversion:", piglatin_word)
```
