```python
#054
import random
mystery = ['h','t'] #create list head and tails
random = random.choice(mystery) #create variable random to collect list
choice = input("Head or Tails (h/t) : ").lower() #ask user to guess and make choice to lowercase
while True :
  if choice == random :
    print("You win")
    break
  else :
    print("Bad luck")
    break
print(f"Computer Selected {random}")
```

```python
#055
import random
random = random.randint(0,5) #create variable to collect random number between 1-5
guess = int(input("Guess Number between 0 to 5 : ")) #ask user to guess number
while True :
  if guess == random :
    print("Well done")
    break
  elif guess != random :
    if guess > random :
      print("Too high")
    elif guess < random :
      print("Too low")
  print("Try again")
  guess2 = int(input("Guess Number between 0 to 5 : "))
  if guess2 == random :
    print("Correct")
  else :
    print("You lose")
    break
```

```python
#058
import random
count = 0

for i in range(5) :
  num1 = random.randint(0,100)
  num2 = random.randint(0,100)
  print(f"Total of {num1} + {num2} = ?")
  total = num1 + num2
  ans = int(input("Enter your answer : "))
  if ans == total :
    count += 1
print(f"Your Score = {count}")
```
